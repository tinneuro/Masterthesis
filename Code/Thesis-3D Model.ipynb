# Params oscillatory regime

```python
w_EE = 1.75 # low recurrence
w_EI = 1.5
w_IE = 2.3
w_II = 0

beta = 0.4

I_E = 4# high external input
I_I = 0

dt = 0.0002

tau_e = 0.01
tau_i = 0.002
tau_a = 1 # both tau_a = 1 and tau_a = 0.5 possible here!


params_sigmoid = [g_e,g_i,theta_e,theta_i,I_E,I_I]
params_interactive = [beta,w_EE,w_EI,w_IE,w_II,g_e,g_i,theta_e,theta_i,I_E,I_I]
params_weights = [beta,w_EE,w_EI,w_IE,w_II]
```

```python
def Sys(X,t,params):
    """The 3D differential equations system to integrate analytically"""
    # X[0] = he, X[1] = hi, X[2] = a
    beta,w_EE,w_EI,w_IE,w_II,g_e,g_i,theta_e,theta_i,I_E,I_I = params
    F_E = nullclines.sigmoidal_transfer(X[0],g_e,theta_e)
    F_I = nullclines.sigmoidal_transfer(X[1],g_i,theta_i)
    he = 1/tau_e*(-X[0] + w_EE*F_E - w_EI*F_I + I_E - X[2])
    hi = 1/tau_i*(-X[1] + w_IE*F_E - w_II*F_I + I_I)
    a = 1/tau_a*(-X[2] + beta*F_E)
    
    return np.array([he,hi,a])
```

```python
# initial condition
Sys0 = np.array([0,0,0])
time = np.arange(0, 10, dt)
```

```python
X, infodict = odeint(Sys,Sys0,time,args = (params_interactive,),full_output=True)
```

```python
he,hi,a = X.T
plt.subplot(2,1,1)
plt.title("Dynamics in time Input Potentials")
plt.plot(time,he,label = "h_e")
plt.plot(time,hi,label = "h_i")
plt.plot(time,a,label = "a",color = "darkorange")
plt.plot(1*np.ones(3),[0,0.02,0.03],color = "black",label = f"timescale of 2*adaptation {2*tau_a} seconds")
plt.xlabel("Seconds")
plt.ylabel("Volt")
#plt.legend(loc = "upper right")

plt.subplot(2,1,2)
plt.title("Dynamics in time rates")
plt.plot(time,model_3D.sigmoidal_transfer(he,g_e,theta_e),label = "r_e")
plt.plot(time,model_3D.sigmoidal_transfer(hi,g_i,theta_i),label = "r_i")
plt.xlabel("Seconds")
plt.ylabel("Hz")
#plt.legend(loc = "upper right")
plt.tight_layout()

plt.show()
```
